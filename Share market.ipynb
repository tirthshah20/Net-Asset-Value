```python
import pandas as pd
```

```python
df=pd.read_csv('NAV1.csv')
```

```python
print(df)
```

```python
df.set_index(df['Date'], inplace=True)
```

```python
df.index =  pd.to_datetime(df.index ,errors='coerce',dayfirst=True)
```

```python
print(df.index)
```

```python
df = df.resample('D').ffill()
```

```python
df.drop('Date',axis='columns',inplace=True)
```

```python
df.rename(columns={'Current value Tata':'NAV Tata','Current Value Mahindra':'NAV Mahindra'},inplace=True)
```

```python
df.head(10)
```

```python
df.shape
```

```python
df['Date']=df.index
```

```python
result = df['Date'].dt.is_month_end
```

```python
df['monthend']=result
```

```python
df.head(10)
```

```python
for i in range(len(df['Adj Close'])):
    if i<=0:
        pass
    else:
        if df['monthend'][i]==True:
            #print(i)
            df['NAV Mahindra'].iloc[i]=(df['Adj Close.1'].iloc[i]/df['Adj Close.1'].iloc[i-1])*df['NAV Mahindra'].iloc[i-1]
            df['NAV Tata'].iloc[i]=(df['Adj Close'].iloc[i]/df['Adj Close'].iloc[i-1])*df['NAV Tata'].iloc[i-1]
            navtata=df['NAV Tata'].iloc[i]
            navmahindra=df['NAV Mahindra'].iloc[i]
            avg=(navtata+navmahindra)/2
            df['NAV Tata'].iloc[i]=avg
            df['NAV Mahindra'].iloc[i]=avg
            
            #df['NAV Mahindra'].iloc[i]=(df['Adj Close'].iloc[i]/df['Adj Close'].iloc[i-1])*df['NAV Mahindra'].iloc[i-1]            
        else:
            
            df['NAV Tata'].iloc[i]=(df['Adj Close'].iloc[i]/df['Adj Close'].iloc[i-1])*df['NAV Tata'].iloc[i-1]
            df['NAV Mahindra'].iloc[i]=(df['Adj Close.1'].iloc[i]/df['Adj Close.1'].iloc[i-1])*df['NAV Mahindra'].iloc[i-1]
            #round(df['Current value Tata'][i])
                
```

```python
df[df['monthend']==True]
```

```python
df['NAV Tata']=round(df['NAV Tata'])
df['NAV Mahindra']=round(df['NAV Mahindra'])
```
